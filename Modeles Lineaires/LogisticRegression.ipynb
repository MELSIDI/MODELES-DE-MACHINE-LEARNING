# LogisticRegression

## Dataset

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2)
```

```python
# Tracé des deux classes séparément pour avoir une légende
plt.figure()
plt.scatter(X_train[y_train == 0, 0], X_train[y_train == 0, 1], c='red', label='Classe 0')
plt.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], c='blue', label='Classe 1')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.grid()
plt.legend()
plt.show()
```

## From Scratch

```python
from LogisticRegression import LogisticRegression
```

```python
model = LogisticRegression(lr=0.1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test, threshold=0.6)
```

```python
accuracy_score(y_test, y_pred)
```

```python
model.fit(X_train[:, : 2], y_train)
```

```python
y_pred = model.predict(X_test[:, : 2])
```

```python
accuracy_score(y_test, y_pred)
```
