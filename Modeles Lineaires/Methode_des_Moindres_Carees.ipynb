# Methode des Equations Normales / Methode des Moindres Carées

## Utilisable avec peux de données

```python
import numpy as np
import matplotlib.pyplot as plt 
```

Generation d'un jeu de données

```python
np.random.seed(1)
x = np.random.randn(100, 1)
y = x / 2 + np.random.randn(100, 1) / 4
```

Representation du jeu de données

```python
plt.figure()
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Observation du jeu de données")
plt.show()
```

Definition du modele

```python
def model(X, theta):
    return X.dot(theta)
```

```python
X = np.hstack((x, np.ones((100, 1)))) # <--- Matrice de features + biais
theta = np.random.randn(2, 1) # <-- Vecteur des coefficient
```

```python
X.shape
```

```python
y.shape
```

```python
theta.shape
```

Répresentation du modele avant l'application de la methode des moindres carrées

```python
y_pred = model(X, theta)
plt.figure()
plt.scatter(x, y, label="Les Observations")
plt.scatter(x, y_pred, c='r',label="Les predictions")
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Observation du jeu de données")
plt.legend()
plt.show()
```

## Application de la methode des moindres carrées

```python
def moindre_carree(X, y):
    A = X.T.dot(X)
    det = np.linalg.det(A)
    if det != 0:
        return np.linalg.inv(A).dot(X.T).dot(y)
    return np.array([[np.nan]])
```

```python
theta = moindre_carree(X, y) # <-- La Methode_des_Moindres_Carees donnes le modele avec la performance la plus optimale 
```

Representation du modele apres l'Application de la methodes des moindres carrrées

```python
if (theta.shape == (1, 1)):
    print("Impossible pas de modele pour ces données")
else:
    x_m = np.linspace(-3, 3, 50).reshape((50, 1))
    y_pred = model(np.hstack((x_m, np.ones((50, 1)))), theta)
    plt.figure()
    plt.scatter(x, y, label="Les Observations")
    plt.plot(x_m, y_pred, c='r',label="Les predictions")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title("Observation du jeu de données")
    plt.legend()
    plt.show()
```

### La performance du modele

```python
def performance(X, theta, y): 
    y_pred = model(X, theta)
    u = ((y - y_pred)**2).sum(axis=0)
    v = ((y - y.mean(axis=0))**2).sum(axis=0)
    return (1 - u / v)[0]
```

```python
print(f"La performance du modele est de {performance(X, theta, y) * 100:.2f} % !!")
```

# Methodes des mondres carrées avec Scikit-Learn

```python
from sklearn.linear_model import LinearRegression # <-- Cette Classe utilise la methodes des equiations Normales
```

```python
model = LinearRegression()
model
```

```python
model.fit(x, y.ravel()) # <-- Entrainement du modele
```

Representation des preductions du modele

```python
x_m = np.linspace(-3, 3, 50).reshape((50, 1))
y_pred = model.predict(x_m)
plt.figure()
plt.scatter(x, y, label="Les Observations")
plt.plot(x_m, y_pred, c='r',label="Les predictions")
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Observation du jeu de données")
plt.legend()
plt.show()
```

```python
print(f"La performance du modele est de {model.score(x, y) * 100:.2f} % !!")
```
