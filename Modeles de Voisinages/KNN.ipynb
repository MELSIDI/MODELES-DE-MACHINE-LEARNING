```python
from sklearn.datasets import load_iris, make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
```

# KNN CLASSIFICATION

## Dataset

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
plt.figure()
plt.scatter(X_train[y_train == 0, 0], X_train[y_train == 0, 1], c='red', label=f"{data.target_names[0]}")
plt.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], c='blue', label=f"{data.target_names[1]}")
plt.scatter(X_train[y_train == 2, 0], X_train[y_train == 2, 1], c='green', label=f"{data.target_names[2]}")
plt.xlabel(f"{data.feature_names[0]}")
plt.ylabel(f"{data.feature_names[1]}")
plt.title("Visualisation")
plt.grid()
plt.legend()
plt.show()
```

## From Scratch

```python
from KNN import KNNClassifier
```

```python
model = KNNClassifier(k=5, weighted=True)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

print(accuracy_score(y_test, y_pred))

## With Sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

# KNN REGRESSION

## Dataset

```python
X, y = make_regression(n_samples=600, n_features=1, noise=30)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
plt.figure()
plt.scatter(X_train, y_train)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Visualisation")
plt.grid()
plt.show()
```

## From Scratch

```python
from KNN import KNNRegressor
from sklearn.metrics import r2_score
```

```python
model = KNNRegressor(k=10, weighted=True, metric="manhattan")
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
```

## With Sklearn

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=10)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
```
