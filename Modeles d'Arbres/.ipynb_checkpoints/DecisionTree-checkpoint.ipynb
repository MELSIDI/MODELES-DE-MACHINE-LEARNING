# DECISION TREE CLASSIFIER

## DATASET

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.title("Visualisation des Données du train Set")
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
```

## FROM SCRATCH

```python
from DecisionTree import DecisionTreeClassifier
```

```python
model_1 = DecisionTreeClassifier(min_samples_split=3, max_depth=100, criterion='entropy')
model_1
```

```python
model_1.fit(X_train, y_train)
```

```python
y_pred = model_1.predict(X_test)
model_1.score(y_test, y_pred)
```

## WITH SKLEARN

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_2 = DecisionTreeClassifier(min_samples_split=3, max_depth=100, criterion='entropy')
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```

# DECISION TREE REGRESSOR

## DATASET

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X = data.data
y = data.target
print("X :", X.shape)
print("y :", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
for i in range(10):
    plt.figure()
    plt.scatter(X_train[:, i], y_train)
    plt.title("DataSet Visialisation")
    plt.xlabel(data.feature_names[i])
    plt.ylabel("Diabete Mesure")
    plt.show()
```

## FROM SCRATCH

```python
from DecisionTree import DecisionTreeRegressor
```

```python
model_1 = DecisionTreeRegressor(min_samples_split=4, max_depth=120)
model_1
```

```python
model_1.fit(X_train, y_train)
```

```python
model_1.score(y_true=y_test, y_pred=model_1.predict(X_test))
```

## WITH SKLEARN

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_2 = DecisionTreeRegressor(min_samples_split=4, max_depth=120)
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```
