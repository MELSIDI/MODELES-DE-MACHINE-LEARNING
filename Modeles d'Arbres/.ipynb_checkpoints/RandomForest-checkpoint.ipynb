# RandomForest

## Classification

### Dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
print("X :", X.shape)
print("y: ", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Visualisation

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.legend()
plt.show()
```

### From Scratch

```python
from RandomForest import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(y_true=y_test, y_pred=model.predict(X_test))
```

### With Sklearn

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_2 = RandomForestClassifier()
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```

## Regression

### Dataset

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_features=30, n_samples=600, noise=0.2)
print("X :", X.shape)
print("y :", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### From Scratch

```python
from RandomForest import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(y_pred=model.predict(X_test), y_true=y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_2 = RandomForestRegressor()
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```
