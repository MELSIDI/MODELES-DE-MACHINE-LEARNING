# DECISION TREE CLASSIFIER

## DATASET

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.title("Visualisation des Donn√©es du train Set")
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
```

## FROM SCRATCH

```python
from DecisionTree import DecisionTreeClassifier
```

```python
model_1 = DecisionTreeClassifier(min_samples_split=3, max_depth=100, criterion='entropy')
model_1
```

```python
model_1.fit(X_train, y_train)
```

```python
y_pred = model_1.predict(X_test)
model_1.score(y_test, y_pred)
```

## WITH SKLEARN

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_2 = DecisionTreeClassifier(min_samples_split=3, max_depth=100, criterion='entropy')
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```

# DECISION TREE REGRESSOR

## DATASET

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=5, noise=10, random_state=2)
print("X :", X.shape)
print("y :", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## FROM SCRATCH

```python
from DecisionTree import DecisionTreeRegressor
```

```python
model_1 = DecisionTreeRegressor(min_samples_split=10, max_depth=1000)
model_1
```

```python
model_1.fit(X_train, y_train)
```

```python
model_1.score(y_true=y_test, y_pred=model_1.predict(X_test))
```

## WITH SKLEARN

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_2 = DecisionTreeRegressor(min_samples_split=10, max_depth=1000)
model_2
```

```python
model_2.fit(X_train, y_train)
```

```python
model_2.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"max_features": range(1, X.shape[1] + 1)}
model = GridSearchCV(DecisionTreeRegressor(), param_grid=param_grid, scoring='r2')
```

```python
model.fit(X_train, y_train,)
```

```python
model.best_score_
```
